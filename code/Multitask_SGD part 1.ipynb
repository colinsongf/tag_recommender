```python
%%px
import nltk, csv, re
from nltk.tokenize import RegexpTokenizer
from bs4 import BeautifulSoup

infilename = "../data/Train_%d.csv" % eid
outfilename = "../data/proc_Train_%d.csv" % eid
import pickle
with open("../working/TOTAL_SIZE.pickle", 'r') as picklefile:
    TOTAL_SIZE = pickle.load(picklefile)
num_eng = 4
split_rows = TOTAL_SIZE / num_eng

# Stop List according to Treebank-POS tags
REJECT = set(["CC", "CD", "DT", "EX", "IN", "LS", "MD", "PDT", "POS",
              "PRP", "PRP$", "RB", "RP", "SYM", "TO", "UH", "WDT", "WP",
              "WP$", "WRB"])
              
ACCEPT = set(["FW", "JJ", "JJR", "JJS", "NN", "NNP", "NNPS", "NNS",
              "RBR", "RBS", "VB", "VBD", "VBG", "VBN", "VBP",
              "VBZ"])
# Additional Stop-words list
sw = set([])
with open("../working/english.stop", 'rb') as f:
    for l in f:
        sw.add(l.rstrip())
# Tokenizer customized for StackOverflow posts
#tokenizer = RegexpTokenizer(r"[^\W_]+[+#]*")
#tokenizer = RegexpTokenizer(r"[^\W\d_-]+['_-]*[^\W\d_-]*[+#]*")
tokenizer = RegexpTokenizer(r"([^\W\d_-]+[+#]*'*[^\W\d_-]*)|([^\W\d_-]+[+#]*[_-]*[^\W\d_-]*)")
# Brill Tagger
tagger = pickle.load(open("../working/treebank_brill_aubt.pickle"))
# WordNet Lemmatizer
lemmatizer = nltk.stem.WordNetLemmatizer()
# Convert treebank POS to WordNet POS
from nltk.corpus.reader.wordnet import ADJ, NOUN, ADV, VERB
wntag_dict = {'J':ADJ, 'V':VERB, 'N':NOUN, 'R': ADV}
def tb2wn(x):
    return wntag_dict[x[0]] if x[0] in wntag_dict else None
```

```python
%%px
def process_title(title):
    # Get rid of appended punctuations
    tokens = tokenizer.tokenize(title)
    # Filter out tokens that have irrelevant POS tags
    tktag_tups = tagger.tag(tokens)
    processed_tups = [(re.sub(r"'+(.*)", '', tup[0]).lower(), tb2wn(tup[1])) for tup in tktag_tups
                     if tup[1] not in REJECT]
    lemmaed_tks = [lemmatizer.lemmatize(t[0], t[1]) for t in processed_tups
                   if ((t[1] is not None) and (t[0] not in sw))]
    concat_tks = ' '.join(lemmaed_tks)
    return concat_tks.encode('ascii', errors='ignore')

def process_body(body):
    # Parse HTML tags
    soup = BeautifulSoup(body)
    body_lst = []
    content_lst = []
    for tag in soup.find_all('p'):
        content_lst.extend(tag.contents)
    for content in content_lst:
        try:
            body_lst.append(content.decode('ascii', errors='ignore')) # <-- beautifulsoup shenanigans
        except TypeError:
            body_lst.append(content.decode(eventual_encoding='ascii')) # <-- beautifulsoup shenanigans
        except:
            body_lst.append(content.encode('ascii', errors='ignore')) # <-- beautifulsoup shenanigans
    tokens_lst = tokenizer.batch_tokenize(body_lst)
    tokens = []
    for tl in tokens_lst:
        tokens.extend(tl)
    # Filter out tokens that have irrelevant POS tags
    tktag_tups = tagger.tag(tokens)
    processed_tups = [(re.sub(r"'+(.*)", '', tup[0]).lower(), tb2wn(tup[1])) for tup in tktag_tups
                     if tup[1] not in REJECT]
    lemmaed_tks = [lemmatizer.lemmatize(t[0], t[1]) for t in processed_tups
                   if ((t[1] is not None) and (t[0] not in sw))]
    concat_tks = ' '.join(lemmaed_tks)
    return concat_tks.encode('ascii', errors='ignore')

def process_csv(infilename=infilename, outfilename=outfilename,
                leftoff=leftoff):
    fin = open(infilename, "rb")
    inreader = csv.reader(fin, delimiter=",")
    ## Ignore header
    #inreader.next();
    
    if leftoff > 0:
        fout = open(outfilename, "ab")
    else:
        fout = open(outfilename, "wb")
    outreader = csv.writer(fout, delimiter=",")
    batch_num = split_rows / 10
    for i,row in enumerate(inreader):
        if i < leftoff: continue
        if (i % batch_num == 0):
            print "Percent completed: %.2f" % (i / batch_num * 10.0)
        newRow = []
        # Pass id
        newRow.append(row[0])
        # Process title
        title = row[1].decode('ascii', 'ignore')
        newTitle = process_title(title)
        newRow.append(newTitle)
        # Process body
        body = row[2].decode('ascii', 'ignore')
        newBody = process_body(body)
        newRow.append(newBody)
        # Pass tags
        newRow.append(row[-1].decode('ascii', 'ignore'))
        # Write to output file
        outreader.writerow(newRow)
    fin.close()
    fout.close()
```

```python
def csr_vappend(a,b):
    """ Takes in 2 csr_matrices and appends the second one to the bottom of the first one. 
    Assumes the type to be csr and overwrites the first matrix instead of copying it.
    The data, indices, and indptr still get copied."""

    a.data = np.hstack((a.data,b.data))
    a.indices = np.hstack((a.indices,b.indices))
    a.indptr = np.hstack((a.indptr,(b.indptr + a.nnz)[1:]))
    a._shape = (a.shape[0]+b.shape[0],b.shape[1])
```

###Hashing and Binarization
Convert `Y` tags-text, `B` body-text, and `T` title-text to hashed sparse format, respectively.

```python
%%px
# Convert pruned2_bodyonly.txt to p=256 scipy.sparse matrix parts
# "Vstack" these matrices on top of another to create B
tagsdict = Dictionary.load("../working/tagsdict.pickle")
with open("../data/tags/pruned2_tagsonly_%d.txt" % eid, 'rb') as f_tags:
    blocks_per_eng = p / num_eng
    for q in xrange(blocks_per_eng):
        if q != blocks_per_eng - 1:
            block_size = split_size / blocks_per_eng
        else: # if last block
            block_size = split_size - (split_size / blocks_per_eng) * (blocks_per_eng - 1)
        
        Y_part = scipy.sparse.lil_matrix((block_size, K), dtype=np.float32)
        for b in xrange(block_size):
            line = f_tags.readline()
            line_set = set()
            for tkn in line.split():
                line_set.add(tagsdict.token2id[tkn])
            for hash_id in line_set:
                Y_part[b, hash_id] = 1
        
        # Initialization
        if q == 0: Y = Y_part.tocsr().copy()
        # Vappend
        else: csr_vappend(Y, Y_part.tocsr())

with open("../data/tags/Y_%d.pickle" % eid, 'w') as picklefile:
    pickle.dump(Y, picklefile)
del Y, tagsdict
```

```python
with open("../data/tags/Y_%d.pickle" % 0, 'r') as picklefile:
    Y = pickle.load(picklefile)

for eid in xrange(1, num_eng):
    with open("../data/tags/Y_%d.pickle" % eid, 'r') as picklefile:
        Y_part = pickle.load(picklefile)
    csr_vappend(Y, Y_part)
del Y_part

with open("../data/tags/Y.pickle", 'w') as picklefile:
    pickle.dump(Y, picklefile)
```
